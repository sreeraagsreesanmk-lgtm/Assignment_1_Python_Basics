```python
# Simple ATM Calculator

balance = int(input("Enter account balance: "))
withdrawal = int(input("Enter withdrawal amount: "))

if withdrawal <= balance:
    balance -= withdrawal
    print(f"Transaction successful. Remaining balance: ₹{balance}")
else:
    print("Insufficient balance. Transaction failed.")
```

```python
# Electricity Bill Generator

units = int(input("Enter units consumed: "))
bill = 0

if units <= 100:
    bill = units * 5
elif units <= 200:
    bill = (100 * 5) + ((units - 100) * 7)
else:
    bill = (100 * 5) + (100 * 7) + ((units - 200) * 10)

print(f"Bill amount = ₹{bill}")
```

```python
# Temperature Converter

temp = float(input("Enter temperature value: "))
unit = input("Enter unit (C/F): ").upper()

if unit == 'C':
    fahrenheit = (temp * 9/5) + 32
    print(f"Temperature in Fahrenheit: {fahrenheit}°F")
elif unit == 'F':
    celsius = (temp - 32) * 5/9
    print(f"Temperature in Celsius: {celsius}°C")
else:
    print("Invalid unit")
```

```python
# Mobile Recharge Validity

from datetime import datetime

recharge_date = input("Enter recharge date (YYYY-MM-DD): ")
today_date = input("Enter today's date (YYYY-MM-DD): ")

recharge = datetime.strptime(recharge_date, "%Y-%m-%d")
today = datetime.strptime(today_date, "%Y-%m-%d")

days_used = (today - recharge).days

if days_used <= 28:
    print("Plan is still valid.")
else:
    print("Plan expired. Recharge again.")
```

```python
# Bus Seat Reservation System

booked = ['Amit', 'Sara', 'John']
cancel = input("Enter name to cancel: ")
new_name = input("Enter new passenger name: ")
position = int(input("Enter position index: "))

if cancel in booked:
    booked.remove(cancel)

booked.insert(position, new_name)
print(booked)
```

```python
# Sum of Even Numbers

n = int(input("Enter a number: "))
even_sum = 0

for i in range(2, n + 1, 2):
    even_sum += i

print("Sum of even numbers:", even_sum)
```

```python
# Student Marks Summary

students = {}
total = 0

for i in range(3):
    name = input("Enter student name: ")
    marks = int(input("Enter marks: "))
    students[name] = marks
    total += marks

average = total / 3
print("Average Marks:", average)
```

```python
# Word Frequency Counter

sentence = input("Enter a sentence: ")
words = sentence.split()

frequency = {}

for word in words:
    frequency[word] = frequency.get(word, 0) + 1

for word, count in frequency.items():
    print(f"{word}: {count}")
```

```python
# Prime Number Checker

def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

num = int(input("Enter a number: "))

if is_prime(num):
    print(f"{num} is a prime number.")
else:
    print(f"{num} is not a prime number.")
```

```python
# Vowel Counter

def count_vowels(text):
    vowels = "aeiouAEIOU"
    count = 0
    for char in text:
        if char in vowels:
            count += 1
    return count

string = input("Enter a string: ")
print("Number of vowels:", count_vowels(string))
```

```python
# Count Divisible Numbers

N = int(input("Enter the range limit (N): "))
D = int(input("Enter the divisor (D): "))

count = 0
for i in range(1, N + 1):
    if i % D == 0:
        count += 1

print(f"Count of numbers divisible by {D}: {count}")
```
